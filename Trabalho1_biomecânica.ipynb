```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas da Primeira Câmera (C1)
print("Informe os Dados da Primeira Câmera (C1)")
x1_c1 = float(input("Informe a coordenada x da primeira câmera (C1): ").replace(',', '.'))
y1_c1 = float(input("Informe a coordenada y da primeira câmera (C1): ").replace(',', '.'))

# Projeção da Primeira Câmera (Projeção 1)
print("Informe os Dados da Projeção da Primeira Câmera (Projeção 1)")
x1_p1 = float(input("Agora, informe a coordenada x da projeção da primeira câmera (Projeção 1): ").replace(',', '.'))
y1_p1 = float(input("Agora, informe a coordenada y da projeção da primeira câmera (Projeção 1): ").replace(',', '.'))

# Coordenadas da Segunda Câmera (C2)
print("Informe os Dados da Segunda Câmera (C2)")
x2_c2 = float(input("Agora, informe a coordenada x da segunda câmera (C2): ").replace(',', '.'))
y2_c2 = float(input("Agora, informe a coordenada y da segunda câmera (C2): ").replace(',', '.'))

# Projeção da Segunda Câmera (Projeção 2)
print("Informe os Dados da Projeção da Segunda Câmera (Projeção 2)")
x2_p2 = float(input("Agora, informe a coordenada x da projeção da segunda câmera (Projeção 2): ").replace(',', '.'))
y2_p2 = float(input("Agora, informe a coordenada y da projeção da segunda câmera (Projeção 2): ").replace(',', '.'))

m1 = (y1_p1 - y1_c1) / (x1_p1 - x1_c1)
m2 = (y2_p2 - y2_c2) / (x2_p2 - x2_c2)

b1 = -1 * (m1 * x1_c1 - y1_c1)  # Coeficiente linear da reta 1
b2 = -1 * (m2 * x2_c2 - y2_c2)  # Coeficiente linear da reta 2

# Ponto de interseção
xi = (-(m2 * x2_c2) + y2_c2 + (m1 * x1_c1) - y1_c1) / (m1 - m2)  # Coordenada x da interseção
yi = m1 * xi - (m1 * x1_c1) + y1_c1  # Coordenada y da interseção

# Exibição das coordenadas do objeto reconstruído
print("Coordenada x do objeto reconstruído:", xi)
print("Coordenada y do objeto reconstruído:", yi)

# Gráfico
x1 = np.linspace(0, 3, 100)
y1 = m1 * x1 + b1

x2 = np.linspace(0, 3, 100)
y2 = m2 * x2 + b2

# Plotagem das retas
plt.plot(x1, y1, color='blue', label='Reta 1')
plt.plot(x2, y2, color='green', label='Reta 2')

# Plotagem do ponto de interseção (objeto reconstruído)
plt.scatter(xi, yi, color='red', label='Objeto Reconstruído')

# Plotagem das posições das câmeras
plt.scatter(x1_c1, y1_c1, color='purple', label='Câmera 1 (C1)')
plt.scatter(x2_c2, y2_c2, color='orange', label='Câmera 2 (C2)')

plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title("Interseção de Retas")
plt.legend()
plt.show()
```
